```python
from typing import Literal
from pydantic import BaseModel
from langchain_google_genai import ChatGoogleGenerativeAI
import os
from dotenv import load_dotenv
load_dotenv()
api_key = os.getenv("GEMINI_API_KEY")

llm=ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite", google_api_key=api_key)


class Intent(BaseModel):
    intent: Literal[
        "browse_menu",
        "find_dishes",
        "ingredients_info",
        "popular_items",
        "place_order",
        "general"
    ]


structured_llm = llm.with_structured_output(Intent)


def detect_intent(user_message: str):

    prompt = f"""
    You are an intent classification system.

    Classify the user's message into exactly one of these intents:

    browse_menu
    - User wants to see menu categories or menu items.

    find_dishes
    - User wants dishes matching preferences such as:
      spicy, veg, chicken, prawn, under ₹500, low calorie, etc.

    ingredients_info
    - User wants ingredients, dietary information,
      calories, veg/non-veg, allergens.

    popular_items
    - User wants recommendations, best sellers,
      trending dishes, chef specials.

    place_order
    - User wants to order, add items,
      modify cart, checkout, place order.

    general
    - Anything else.

    User message:
    {user_message}

    Return only the intent.
    """

    result = structured_llm.invoke(prompt)

    return result.intent
```
