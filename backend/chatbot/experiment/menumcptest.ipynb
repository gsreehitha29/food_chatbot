```python
from fastmcp import FastMCP
mcp = FastMCP("RestaurantMenuMCP")
```

```python
from pymongo import MongoClient

client = MongoClient("mongodb+srv://Cluster47062:XLxGVu26VDbfppvU@cluster47062.yhvoyo7.mongodb.net/")
db = client.chatbot
menu = db.menu
print(menu.find_one())
```

```python
@mcp.tool()
def get_categories():
    """
    Returns all available menu categories
    """
    return list(menu.distinct("category"))
```

```python
get_categories()
```

```python
@mcp.tool()
def get_menu_items(category: str):
    """
    Returns all items in a given category
    """
    return list(menu.find({"category": category}))
```

```python
get_menu_items('Main Course')
```

```python
@mcp.tool()
def search_menu(query: str):
    """
    Search menu items by name (case-insensitive)
    """

    return list(
        menu.find(
            {
                "item_name": {
                    "$regex": query,
                    "$options": "i"
                }
            },
            {"_id": 0}
        )
    )
```

```python
search_menu('Prawn')
```

```python
@mcp.tool()
def get_item_details(item_name: str):
    """
    Get exact item details from MongoDB menu collection
    """

    item = menu.find_one(
        {
            "item_name": {
                "$regex": f"^{item_name}$",
                "$options": "i"
            }
        },
        {
            "_id": 0
        }
    )

    if not item:
        return {
            "error": "Item not found"
        }

    return item
```

```python
get_item_details('Prawn Curry')
```
